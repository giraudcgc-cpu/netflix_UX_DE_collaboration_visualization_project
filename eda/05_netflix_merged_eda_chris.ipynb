### Cleaning country code

```python
# Cleaning country_iso2
df = duckdb.sql("""---sql
SELECT *
    RENAME (country_iso2 AS country_code) 
FROM 
    df
""").df()
df
```

### Cleaning runtime to round it

```python
# NaN --> 0 | rounding | float --> integer
df['runtime'] = df['runtime'].fillna(0).round(0).astype(int)
df.head(2)
```

### Creating Average view in min avg_view_min (KPI)

```python
# Using global columns for metrics like in perf marketing with avg engag. time (possibel KPI?)
#  Creatting a new column + exchanging Nan --> 0 + convert float --> integer
df['avg_view_min'] = ((df['global_weekly_hours_viewed'] * 60) / df['global_weekly_views']).fillna(0).round(0).astype(int)
df.head(1)
```

### Creating a complete watch rate complete_watch_rate

```python
# Complete watch ratio to see the film portion that people saw / maybe better than avg_view_min because films have diff length (KPI?)
df['complete_watch_rate'] = df['avg_view_min'] / df['runtime']
df.head(2)
# Note: complete_watch_rate peut etre > a 1 ou 100%, erreur ou ok? = Kokchun | explanation: user can do backwards to watch somehting else or cut at 1 ?
```

```python
# Test of the above, by asc order to check if works
#df_sorted = df.sort_values(by=['complete_watch_ratio', 'avg_view_min'], ascending=False
duckdb.sql("""---sql
    SELECT 
        show_title as film,
        runtime,
        avg_view_min,
        complete_watch_rate
    FROM
        df
    ORDER BY 
        complete_watch_rate
""").df()
```

```python
# Excluding when avg_view_min is == 0 and runtime == 0 as impossible
duckdb.sql("""
    SELECT 
        show_title as film,
        runtime,
        avg_view_min,
        complete_watch_rate
    FROM
        df
    WHERE 
        avg_view_min > 0
        AND runtime > 0
    ORDER BY 
        complete_watch_rate ASC
""").df().head(20)
# repetition comes the series and diff seasons
```
